```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data = pd.read_csv('SatSex22.csv', encoding='UTF-8', sep=';')
data.iloc[9,5] = 11.53
data.iloc[1,8] = 23
data["Режим Работы"] = data["Режим Работы"].astype('float64')
data
```

```python
data.dtypes
```

```python
man_data = data[(data["Пол"]) == 1]
woman_data = data[(data["Пол"]) == 2]
all_data = data[(data["Пол"]) == 0]
```

```python
all_data
```

```python
fig, axs = plt.subplots(nrows= 1 , ncols= 2)
plt.figure(figsize=(1,2))
data1 = all_data["Заработок"]
labels = all_data["Удовлетворённость"]

data2 = man_data["Заработок"]


data3 = woman_data["Заработок"]

colors = sns.color_palette('ch:start=.2,rot=-.3')

axs[1].pie(data2, labels = labels, colors = colors, autopct='%.0f%%')
axs[1].set_title("Men")
axs[2].pie(data3, labels = labels, colors = colors, autopct='%.0f%%')
axs[2].set_title("Women")


plt.show()
```

Несложно заметить, что удовлетворенность заработной платой у женщин меньше, чем у мужчин на 5%. При этом, во всех остальных параметрах удовлетворенности женщины лидируют.
